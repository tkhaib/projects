<b>Now let us return to the issue of more than 6000 entries being Nan for 28 variables and more than 2000 entries being Nan for 4 variables. It is easy to see that there are mostly exatly the same amount of missing values for several variables. Let us double check that before taking any actions.</b>

```python
import pandas as pd
import itertools

col_pairs = list(itertools.combinations(df_qualitative.columns, 2))

for col1, col2 in col_pairs:
    missing1 = df[col1].isnull().sum() / df.shape[0] * 100 
    missing2 = df[col2].isnull().sum() / df.shape[0] * 100
    if missing1 > 60 and missing2 > 60:
        both_null = df[col1].isnull() & df[col2].isnull()
        count = both_null.sum()
        print(f'Columns: {col1}, {col2} - Overlapping 60% Nan: {count}')

    if missing1 > 20 and missing1 < 30 and missing2 > 20 and missing2 < 30:
        both_null = df[col1].isnull() & df[col2].isnull()
        count = both_null.sum()
        print(f'Columns: {col1}, {col2} - Overlapping 20% Nan: {count}')
```

<b> As it now can clearly be seen, most of the variables have their missed values in the same rows. However, when it comes to the qualitative features, the such amount of data cannot be restored, thus we'd have to remove it. The quantitative features are arguable to remove, as there are several methods to try to predict the values instead of just cutting it out, but with missing over 50% of the data, we cannot claim that those predictions would even be reliable at all. Thus, after making sure that most of the missing values overlap for quantitative features too, we will get to the amputation procedure. </b>

<b>Also we've found out that there are two variables among the 4 that we've highlighted previously (that had roughly 20% of missing values), and that have most of their lost values intersect, but we will leave them for now. Applying the same procedure for quantitative features: </b>

```python
col_pairs = list(itertools.combinations(df_quantitative.columns, 2))

for col1, col2 in col_pairs:
    missing1 = df[col1].isnull().sum() / df.shape[0] * 100 
    missing2 = df[col2].isnull().sum() / df.shape[0] * 100
    if missing1 > 60 and missing2 > 60:
        both_null = df[col1].isnull() & df[col2].isnull()
        count = both_null.sum()
        print(f'Columns: {col1}, {col2} - Overlapping 60% Nan: {count}')

    if missing1 > 20 and missing1 < 30 and missing2 > 20 and missing2 < 30:
        both_null = df[col1].isnull() & df[col2].isnull()
        count = both_null.sum()
        print(f'Columns: {col1}, {col2} - Overlapping 20% Nan: {count}')
```

<b>Now we can successfully remove all of the unnessasary variables. However we will require the "sum_of_paym_months", "min_MnthAfterLoan" and "all_credits" for the RFM analysis as those variables will be taken as the "Monetary", "Recency", and "Frequency" respectively. Unfortunaly more than 50% of the data is missed, but we will fill the Nan values with the mean values of each column, as there are no better options.</b>

```python
rfm = ["sum_of_paym_months", "min_MnthAfterLoan", "all_credits"]
quantitative = list(filter(lambda elem: elem not in rfm, quantitative))
df.drop(new_df[new_df['%Null'] > 60].filter(items=quantitative, axis=0).index.tolist(), axis=1, inplace=True)
quantitative = list(set(quantitative) - set(new_df[new_df["%Null"] > 60].filter(items=quantitative, axis=0).index.tolist())) + rfm
df.drop(new_df[new_df['%Null'] > 60].filter(items=qualitative, axis=0).index.tolist(), axis=1, inplace=True)
qualitative = list(set(qualitative) - set(new_df[new_df["%Null"] > 60].filter(items=qualitative, axis=0).index.tolist()))
new_df2 = pd.concat([df.nunique().rename('Unique'), (df.nunique() / df.shape[0] * 100).rename('%Unique'), df.isnull().sum().rename('Null'), (df.isnull().sum() / df.shape[0] * 100).rename('%Null'), (df == 0).sum().rename('Zero'), ((df == 0).sum() / df.shape[0] * 100).rename('%Zero')], axis=1)
new_df2
```

<b>We have successfully removed the part of the data, which possessed too much missing values. ## Part 2: Segmentation

### KMeans

<b> First, let us decide on how many clusters should we choose. But first, lets decrease the dimensionality of the data, using the Principal components analysis (PCA) and Scale our data in order for some of the variables not to outweight other variables. For simplicity and understandability pipeline is used.</b>

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

X = df_filled.values

pipeline = Pipeline([
    ('scaling', MinMaxScaler()), 
    ('pca', PCA(n_components=15))
])

X_transformed = pipeline.fit_transform(X)
df_transformed = pd.DataFrame(X_transformed)

wcss = [] 

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

<b> As it can be seen, after the Elbow method implementation it is reasonable to choose 6 clusters. Choosing 4 clusters might also seem to be optimal, but there are risks of low detlisation. </b>

```python
clusterise = KMeans(n_clusters=6).fit_predict(X)
df_forlater['Cluster'] = clusterise
df_forlater.head(10)
```
